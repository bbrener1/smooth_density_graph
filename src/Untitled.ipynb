```python
import smooth_density_field as sdf
import numpy as np
import matplotlib.pyplot as plt
```

```python
counts = np.loadtxt('../../reborrow_clustering/encoding.txt')
# counts = np.loadtxt('../../reborrow_clustering/iris.trunc')
```

```python
# nm = sdf.neighborhood_models(counts,neighbors=5,models=10,sample_fraction=.2)
```

```python
final_pos,fuzz = sdf.converge(counts,sample_fraction=.05,models=10)
```

```python
clusters = sdf.quick_cluster(final_points=final_pos,fuzz=fuzz)
```

```python
cell_order = np.argsort(clusters)
```

```python
for cluster in set(clusters):
    print(np.sum(np.array(clusters) == cluster))
```

```python
plt.figure()
plt.imshow(counts,aspect='auto')
plt.show()
```

```python
plt.figure()
plt.imshow(counts[cell_order],aspect='auto')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
# d_cell_order = dendrogram(linkage(counts,metric='cos',method='average'),no_plot=True)['leaves']
leaf_order = dendrogram(linkage(counts.T,metric='cos',method='average'),no_plot=True)['leaves']

plt.figure()
plt.imshow(counts[cell_order].T[leaf_order].T,aspect='auto')
plt.show()
```
